```python
def mul_img_fnd(dir_list,low_threshold,high_threshold):#finds multiple images folders in a given superior directory
  multiple_img = []
  for sf in dir_list:
    sf_path = os.path.join(main_folder_path, f"{sf}")#joins dir ath with subfolders
    cnt=0
    for j in os.listdir(sf_path):
      cnt+=1
    if low_threshold < cnt and cnt < high_threshold:#gets folders with no. of images between high and low thresholds
      multiple_img.append(sf)
  return multiple_img
```

```python
def get_paths_cnt(dir_list1,main_folder_path,strt_indx,end_indx):#gets paths of all the folders and images in them provided length of directory
  x=[]
  y = []
  cnt=0
  if len(dir_list1)==0:
    dir_list1 = os.listdir(main_folder_path)
  for sf in dir_list1[strt_indx:end_indx]:
    sub_paths = os.path.join(main_folder_path, f"{sf}")
    y.append(sub_paths)
    for ssf in os.listdir(sub_paths):
      x.append(os.path.join(sub_paths,f"{ssf}"))
      cnt+=1
  return x,y,cnt
```

```python
def get_max_cnt(dir_list1):#getting max count from root folder
  count=[]
  for sf in dir_list1[:]:
    cnt=0
    sub_paths = os.path.join(main_folder_path, f"{sf}")
    for ssf in os.listdir(sub_paths):
      cnt+=1
    count.append(cnt)
  return max(count)
```

```python
from PIL import Image
def extract_name_from_path(file_path):
    path_parts = file_path.split('/')
    name = path_parts[-2]
    return name

train_paths = get_paths_cnt(mul_img_fnd(train_set,1,53),main_folder_path,0,len(mul_img_fnd(train_set,1,53)))[0]#make ttrain images path array
val_paths = get_paths_cnt(mul_img_fnd(val_set,1,53),main_folder_path,0,len(mul_img_fnd(val_set,1,53)))[0]#make val images path array
test_paths = get_paths_cnt(mul_img_fnd(test_set,1,53),main_folder_path,0,len(mul_img_fnd(test_set,1,53)))[0]#make test images path array


random.shuffle(train_paths)
random.shuffle(val_paths)

train_dir_path = "/content/train_dir"
val_dir_path = "/content/val_dir"
test_dir_path = "/content/test_dir"
os.mkdir(train_dir_path)
os.mkdir(val_dir_path)#making different directories
os.mkdir(test_dir_path)

for img in train_paths:
    img_dir = extract_name_from_path(img)#extracting name of folder from image paths
    if not os.path.isdir(os.path.join(train_dir_path,img_dir)) :
      os.mkdir(os.path.join(train_dir_path,img_dir))
    shutil.copy(img, os.path.join(train_dir_path,img_dir))#copying folder from root dir to specific dir

for img in val_paths:
    img_dir = extract_name_from_path(img)
    if not os.path.isdir(os.path.join(val_dir_path,img_dir)):
      os.mkdir(os.path.join(val_dir_path,img_dir))
    shutil.copy(img, os.path.join(val_dir_path,img_dir))

for img in test_paths:
    img_dir = extract_name_from_path(img)
    if not os.path.isdir(os.path.join(test_dir_path,img_dir)):
      os.mkdir(os.path.join(test_dir_path,img_dir))
    shutil.copy(img, os.path.join(test_dir_path,img_dir))

```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import cv2
from google.colab.patches import cv2_imshow


# Define a function to create a Siamese data generator
# Define a function to create a Siamese data generator
def siamese_data_generator(generator):
    m=0
    while m<=1:
        pairs = []
        labels = []
        print(len(generator))
        for _ in range(len(generator)):
            x, y = next(generator)
            for i in range(len(y)):
                for j in range(i+1, len(y)):
                    pairs.append([x[i], x[j]])  # Create pairs of images
                    labels.append(1 if y[i] == y[j] else 0)  # 1 if from the same class, else 0
        print(np.array(labels))
        m+=1
        yield [np.array(pairs)[:, 0], np.array(pairs)[:, 1]], np.array(labels)  # Return pairs as two separate arrays, one for each image

# Load the pre-trained Siamese model
model = tf.keras.models.load_model('/content/drive/MyDrive/best_siamese_model.h5')

# Create an ImageDataGenerator for test data
datagen = ImageDataGenerator(rescale=1./255)
test_generator = datagen.flow_from_directory(
    '/content/test_dir',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False  # Set this to False to maintain pair order
)

# Create a Siamese data generator for testing
test_siamese_generator = siamese_data_generator(test_generator)

# Make predictions using the pre-trained Siamese model
predictions = model.predict(test_siamese_generator, steps=len(test_generator), verbose=1)

# 'predictions' will contain the predicted similarity scores for each pair

```

```python
import cv2

# Iterate through all pairs of images
k=0
for i in range(19):
    image_path1 = test_generator.filepaths[i]
    label1 = test_generator.labels[i]

    for j in range(i + 1, 19):
        label2 = test_generator.labels[j]
        image_path2 = test_generator.filepaths[j]

        if predictions[k] >= 0.5 and label1 == label2:  # Adjust this threshold based on your problem
            label_text = "Same Person and predicted as same"
            color = (0, 255, 0)  # Green for same
        elif predictions[k] < 0.5 and label1 != label2:
            label_text = "Different People and predicted different"
            color = (0, 0, 255)  # Red for different
        elif predictions[k] < 0.5 and label1 == label2:
            label_text = "Same Person but predicted different"
            color = (255, 0, 0)  # blue for incorrect
        elif predictions[k] >= 0.5 and label1 != label2:
            label_text = "Different person but predicted same"
            color = (255, 0, 0)  #blue for incorrect
        k+=1
        img1 = cv2.imread(image_path1)
        img2 = cv2.imread(image_path2)

        img1 = cv2.putText(img1, label_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
        img2 = cv2.putText(img2, label_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

        # Concatenate the two images side by side
        concatenated_img = cv2.hconcat([img1, img2])

        cv2_imshow(concatenated_img)
```
